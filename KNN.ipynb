```python
from typing import List
import numpy as np
from vector_space_model import VectorSpaceModel
```

```python
vector_space_model = VectorSpaceModel(
    documents_path="ResearchPapers",
    stop_words_file_path="Stopword-List.txt",
)
vector_space_model.initiate()
```

```python
X = []
doc_seq = []
for i in vector_space_model.vsm_index:
    doc_seq.append(int(i[0]))
    X.append(i[1])
y = vector_space_model.make_y(doc_seq=doc_seq)
X = np.array(X)


# from sklearn.preprocessing import Normalizer

# transformer = Normalizer().fit(X)  # fit does nothing.
# B = transformer.transform(X)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.48, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Determine the number of splits for cross-validation
num_splits = 5

# Define the KNN model
knn = KNeighborsClassifier()

# Define the grid of hyperparameters to search
param_grid = {
    # 'n_neighbors': np.arange(1, 6),  # Search for neighbors from 1 to 5
    'weights': ['uniform', 'distance'],  # Search for uniform and distance weights
    'metric': ['euclidean', 'manhattan', 'minkowski']  # Search for different distance metrics
}

# Perform grid search using GridSearchCV with stratified k-fold cross-validation
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=StratifiedKFold(n_splits=num_splits), scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Instantiate a new KNN model with the best parameters
best_knn = KNeighborsClassifier(**best_params)

# Fit the model on the entire training set with best hyperparameters
best_knn.fit(X_train_scaled, y_train)

# Evaluate the model
score = best_knn.score(X_test_scaled, y_test)
print("Accuracy:", score)

# Make predictions
y_pred = best_knn.predict(X_test_scaled)

# Generate confusion matrix and classification report
matrix = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
print("Confusion Matrix:\n", matrix)
print("Classification Report:\n", report)
```
